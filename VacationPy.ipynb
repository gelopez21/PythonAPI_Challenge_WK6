### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#Configured API (g_key)
gmaps.configure(api_key=g_key)
locations=citydata_df[["Lat","Lng"]]
humidity=citydata_df["Humidity"]
```

```python
humidityfig = gmaps.figure()
heat_layer=gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, point_radius=1)

humidityfig.add_layer(heat_layer)

humidityfig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
fitweathercities_df = citydata_df[(citydata_df["Humidity"] > 70) & (citydata_df["Humidity"] <= 75)]
fitweathercities_df = fitweathercities_df.dropna()

del fitweathercities_df['Unnamed: 0']

fitweathercities_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
#fitweathercities dataframe as hotel_df
hotel_df=fitweathercities_df.reset_index(drop=True)

#"Hotel Name" column to dataframe
hotel_df["Hotel Name"] = ""

hotel_df.reset_index(inplace=True,drop=True)

hotel_df.head()
```

```python
#fitweathercities dataframe as hotel_df
hotel_df = fitweathercities_df.astype(str)
#"Hotel Name" column to dataframe
hotel_df["Hotel Name"] = ""

#Url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?parameters"

#hotels within 5000 meters
for index,row in hotel_df.iterrows():
    try:
        location = row[str("Lat")] + "," + row[str("Lng")]
        request_url = base_url + "location=" + location + "&radius=5000" + "&type=lodging" + "&key=" + g_key
        response = requests.get(request_url).json()
        hotel_name = response["results"][0]["name"]
        row["Hotel Name"] = hotel_name
    except IndexError:
        row["Hotel Name"] = "**No hotel within 5000 meters**"
        
hotel_df.head(7)
```
